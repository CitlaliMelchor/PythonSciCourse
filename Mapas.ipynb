```python
import folium
import pandas as pd
 
# Make a data frame with dots to show on the map
df = pd.read_csv("starbucks.csv")

df.head()
```

```python
list(set(df["City"]))[:5]
```

```python
df = df.query("City=='Mexico DF' or City=='México DF'")

df.head()
```

```python
df.columns
```

```python
df = pd.DataFrame(df, columns=["Store Number", "Longitude", "Latitude"])

df.head()
```

```python
m = folium.Map(location=[19.41, -99.18], tiles="Stamen Toner", zoom_start=12)
 
# I can add marker one by one on the map
for i in range(0, len(df)):
    folium.Marker([df.iloc[i]['Latitude'], df.iloc[i]['Longitude']], popup=df.iloc[i]['Store Number']).add_to(m)
    
m
```
