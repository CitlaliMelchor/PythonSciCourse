### Ejemplo 1 : líneas y marcadores
El siguiente código muestra una gráfica sencilla que se muestra directamente aquí, pero si el código se ejectuta en una terminal, abrirá una ventana nueva y ahí mostrará la gráfica.

```python
import matplotlib.pyplot as plt #importación por alias
x=[1,2,3,4,5,6]
y=[5,3,6,1,2,4]

plt.plot(x,y)
plt.show()
```

Las gráficas pueden ser configuradas en color y forma mediante parámetros adiciones. Por ejemplo, los siguientes * **marcadores** * son muy útiles, para ver la lista completa de marcadores visita: [https://matplotlib.org/api/markers_api.html](URL "Marcadores")
* `o` - círculo
* `+` - signo de más
* `*` - estrella 
* `s` - cuadrado
* `<, >, ^, v` - triángulos orientados  

Los marcadores pueden ser combinados con líneas `-`, líneas punteadas `--` o sin línea. 

La librería también soporta colores nombrados y exadecimales, un generador de colores ese encuentra en [https://www.w3schools.com/colors/colors_picker.asp ](URL "color picker")

El siguiente código muestra cómo dibujar varias seres con diferentes marcadores y colores

```python
import matplotlib.pyplot as plt
x1=[1,2,3]
y1=[1,2,3]

x2=[1,2,3]
y2=[3,1,2]

x3=[1,3,5]
y3=[2,4,6]
plt.plot(x1,y1, "o", color="hotpink")
plt.plot (x2,y2, "^-", color="lime")
plt.plot(x3,y3, "s--", color="cornflowerblue")
plt.show()
```

### Ejemplo 2: Gráfica de pastel
* Sizes/Weights - Elementos a ponderar el pastel
* Labels - Etiquetas de cada categoría
* Explode - Separación del pastel
* Autopctg - Formato del porcentaje

```python
import matplotlib.pyplot as plt 
sizes = [3,4,2,7,5]
labels =["manzanas","peras", "mangos", "kiwis", "plátanos"] 
explode =[0, 0, 0.1, 0, 0]
autopct="%.2f%%" # C - %x.yy
colors=["red","lime", "gold", "olive", "Coral"]


plt.pie(sizes, labels=labels, explode=explode, autopct=autopct, colors=colors)
plt.axis("equal")
plt.show()
```

### Ejemplo 3: Gráfica de barras

* Positions - Posiciones de las barras
* Heights - Alturas de las barras
* Weights - Ancho de las barras

```python
import matplotlib.pyplot as plt
x= [1,2,3,4]
y=[10,8,5,9]
weight=0.2

plt.bar(x,y,weight, color="plum", tick_label=["A", "B", "C", "D"])

for i, j in zip(x,y):
    plt.text(i,j,j, ha="center", va="bottom")
    
plt.show()
```
